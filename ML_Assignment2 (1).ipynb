```python
import pandas as pd
import numpy as np

class NaiveBayes():
    def __init__(self, train_df, test_df):
        # Initializing the training and testing datasets
        self.training_dataset = train_df
        self.testing_dataset = test_df

        self.likelihood = {}
        features = list(self.training_dataset.columns)[:-1]

        for f in features:
            self.likelihood[f] = {}
            for v in self.training_dataset[f].unique():
                self.likelihood[f][v] = {}
                for i in np.unique(self.training_dataset['class']):
                    self.likelihood[f][v][i] = 1e-3

        self.prior_probability = []

        self.income_classes = np.unique(self.training_dataset['class'])
        for i in self.income_classes:
            self.prior_probability.append(len(self.training_dataset[self.training_dataset['class'] == i]) / len(self.training_dataset))

    def calculate_likelihood(self):
        features = list(self.training_dataset.columns)[:-1]
        for i in np.unique(self.training_dataset['class']):
            data = self.training_dataset[self.training_dataset['class'] == i]
            for f in features:
                for v in self.training_dataset[f].unique():
                    if v in data[f].values:
                        count = data[f].value_counts()[v] + 1e-3
                    else:
                        count = 1e-3
                    self.likelihood[f][v][i] = count / data.shape[0]

    def naive_bayes(self):
        self.calculate_likelihood()
        testing_data = self.testing_dataset.iloc[:, :-1]
        predictions = []
        for i in range(len(testing_data)):
            posterior_probabilities = []
            for j in self.income_classes:
                posterior_probability = self.prior_probability[j]
                for f in testing_data.columns:
                    posterior_probability *= self.likelihood[f][testing_data.iloc[i][f]][j]
                posterior_probabilities.append(posterior_probability)
            predictions.append(self.income_classes[np.argmax(posterior_probabilities)])
        return predictions


# Example usage
# Assume you have train_df and test_df as your datasets
nb = NaiveBayes(train_df, test_df)
y_pred = nb.naive_bayes()
# print("Predicted labels:", y_pred)

# Calculate and print the confusion matrix
def confusion_matrix(y_pred, y_real):
    matrix = {
        'TP': 0,
        'TN': 0,
        'FN': 0,
        'FP': 0,
    }

    for i, j in zip(y_pred, y_real):
        if i == 0 and j == 0:
            matrix["TN"] += 1
        elif i == 0 and j == 1:
            matrix["FN"] += 1
        elif i == 1 and j == 0:
            matrix["FP"] += 1
        elif i == 1 and j == 1:
            matrix["TP"] += 1

    print("Confusion Matrix:")
    print("TN:", matrix["TN"])
    print("FN:", matrix["FN"])
    print("FP:", matrix["FP"])
    print("TP:", matrix["TP"])

confusion_matrix(y_pred, test_df['class'])




```

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score

# Define the list of optimizers
optimizers = ['adam', 'sgd', 'rmsprop']
num_epochs = range(1, 11)

# Initialize lists to store accuracy values
accuracies = []

# Iterate over all optimizers
for optimizer in optimizers:
    # Define the neural network
    model = Sequential()
    model.add(Dense(150, input_shape=(X_train.shape[1],)))
    model.add(Dense(100, input_shape=(X_train.shape[1],)))
    model.add(Dense(100, input_shape=(X_train.shape[1],)))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model with the current optimizer
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model and evaluate on the test data for each number of epochs
    epoch_accuracies = []
    for num_epoch in num_epochs:
        model.fit(X_train, y_train, epochs=num_epoch, batch_size=32, verbose=0)
        y_pred_prob = model.predict(X_test)
        y_pred = np.where(y_pred_prob > 0.5, 1, 0)  # Convert probabilities to classes based on a threshold
        accuracy = accuracy_score(y_test, y_pred)
        epoch_accuracies.append(accuracy)

    accuracies.append(epoch_accuracies)

# Convert the accuracies list to a numpy array
accuracies = np.array(accuracies)

# Plot the graph
fig, ax = plt.subplots()

# Iterate over the optimizers and plot the accuracies
for i, optimizer in enumerate(optimizers):
    ax.plot(num_epochs, accuracies[i], marker='o', label=optimizer)

ax.set_xlabel('Number of Epochs')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy vs. Number of Epochs')
ax.legend()

plt.show()
```

# checking the accuracy for different type of optimizers.

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score

# Define the list of optimizers
optimizers = ['adam', 'sgd', 'rmsprop']
num_epochs = range(1, 11)

# Initialize lists to store accuracy values
accuracies = []

# Iterate over all optimizers
for optimizer in optimizers:
    # Define the neural network
    model = Sequential()
    model.add(Dense(150, input_shape=(X_train.shape[1],)))
    model.add(Dense(100, input_shape=(X_train.shape[1],)))
    model.add(Dense(100, input_shape=(X_train.shape[1],)))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model with the current optimizer
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model and evaluate on the test data for each number of epochs
    epoch_accuracies = []
    for num_epoch in num_epochs:
        model.fit(X_train, y_train, epochs=num_epoch, batch_size=32, verbose=0)
        y_pred_prob = model.predict(X_test)
        y_pred = np.where(y_pred_prob > 0.5, 1, 0)  # Convert probabilities to classes based on a threshold
        accuracy = accuracy_score(y_test, y_pred)
        epoch_accuracies.append(accuracy)

    accuracies.append(epoch_accuracies)

# Convert the accuracies list to a numpy array
accuracies = np.array(accuracies)

# Plot the graph
fig, ax = plt.subplots()

# Iterate over the optimizers and plot the accuracies
for i, optimizer in enumerate(optimizers):
    ax.plot(num_epochs, accuracies[i], marker='o', label=optimizer)

ax.set_xlabel('Number of Epochs')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy vs. Number of Epochs')
ax.legend()

plt.show()
```

# We will now analyze the accuracy of the model using different activation functions while keeping the optimizer constant as "adam." This choice is based on the observation from the previous graph, which indicated that the "adam" optimizer generally provided the best accuracy across different activation functions.

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score

# Define the optimizer and activation functions
optimizer = 'adam'
activations = ['relu', 'sigmoid', 'tanh']
num_epochs = range(1, 11)

# Initialize lists to store accuracy values
accuracies = []

# Iterate over all activation functions
for activation in activations:
    # Define the neural network
    model = Sequential()
    model.add(Dense(150, activation=activation, input_shape=(X_train.shape[1],)))
    model.add(Dense(100, activation=activation))
    model.add(Dense(100, activation=activation))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model with the specified optimizer
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model and evaluate on the test data for each number of epochs
    epoch_accuracies = []
    for num_epoch in num_epochs:
        model.fit(X_train, y_train, epochs=num_epoch, batch_size=32, verbose=0)
        y_pred_prob = model.predict(X_test)
        y_pred = np.where(y_pred_prob > 0.5, 1, 0)  # Convert probabilities to classes based on a threshold
        accuracy = accuracy_score(y_test, y_pred)
        epoch_accuracies.append(accuracy)

    accuracies.append(epoch_accuracies)

# Convert the accuracies list to a numpy array
accuracies = np.array(accuracies)

# Plot the accuracies for each activation function
plt.figure(figsize=(10, 6))
for i, activation in enumerate(activations):
    plt.plot(num_epochs, accuracies[i], marker='o', label=activation)

plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Number of Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Print the accuracy for each activation function
for i, activation in enumerate(activations):
    print(f"Accuracy for Activation Function {activation}: {accuracies[i][-1]}")
```

# We will now visualize the accuracy vs. epochs graph for different numbers of layers for different size of epochs, with the optimizer set to "adam" and the activation function set to "sigmoid". This combination of optimizer and activation function showed the best accuracy performance according to the previous graph.

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score

# Define the number of layers and epochs
num_layers = [1, 2, 3]
num_epochs = range(1, 21)

# Initialize lists to store accuracy values for each number of layers
accuracies = [[] for _ in num_layers]

# Iterate over the number of layers
for i, num_layer in enumerate(num_layers):
    # Iterate over the number of epochs
    for num_epoch in num_epochs:
        # Define the neural network
        model = Sequential()
        model.add(Dense(100, activation='sigmoid', input_shape=(X_train.shape[1],)))
    
        # Add additional hidden layers
        for _ in range(num_layer - 1):
            model.add(Dense(100, activation='sigmoid'))
    
        model.add(Dense(1, activation='sigmoid'))
    
        # Compile the model
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
        # Train the model
        model.fit(X_train, y_train, epochs=num_epoch, batch_size=32, verbose=0)
    
        # Evaluate the model on the test data
        _, accuracy = model.evaluate(X_test, y_test, verbose=0)
    
        # Append the accuracy to the accuracies list
        accuracies[i].append(accuracy)

# Plot the graph
for i, num_layer in enumerate(num_layers):
    plt.plot(num_epochs, accuracies[i], marker='o', label=f'{num_layer} Layer(s)')

plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Number of Epochs for Different Number of Layers')
plt.legend()
plt.show()
```
