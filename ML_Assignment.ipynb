```python
import docx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the Word document
doc = docx.Document('census-income.data.docx')

# Extract the content from the Word document
content = []
for paragraph in doc.paragraphs:
    content.append(paragraph.text)

# Remove empty rows from content
content = [row for row in content if row.strip()]

# Define column names
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','class']

# Create a dictionary of column names and content
data = {}
for i, col_name in enumerate(column_names):
    data[col_name] = [row.split(',')[i].strip() if len(row.split(',')) > i else '' for row in content]

# Create a DataFrame from the dictionary
train_df = pd.DataFrame(data)

# Define the output CSV file path
csv_file_path = 'train.csv'

# Write the DataFrame to a CSV file
train_df.to_csv(csv_file_path, index=False)

# Read the CSV file
train_df = pd.read_csv(csv_file_path)

print("CSV file created successfully!")
```

```python
# Calculate the count of "?" values in each column
question_mark_counts = (train_df == "?").sum()

# Display the count of "?" values for each column
# print("Question Mark Counts:")
# print(question_mark_counts)
# Replace "?" with NULL
train_df.replace("?", np.nan, inplace=True)

# Calculate the count of NULL values in each column
null_value_counts = train_df.isnull().sum()

# Display the count of NULL values for each column
print("NULL Value Counts:")
print(null_value_counts)
```

```python
# Display the data statistics
# Calculate the statistics for each column
data_stats = train_df.describe(include='all')
print("Data Statistics:")
print(data_stats)
```

```python
# Calculate the count of NULL values in each column
null_value_counts = train_df.isnull().sum()

# Calculate the percentage of NULL values per column
null_value_percentages = (null_value_counts / len(train_df)) * 100

# Display the null count percentages per column
print("Null Count Percentages per Column:")
print(null_value_percentages)
```

```python
# Read the Word document
doc = docx.Document('census-income.test.docx')

# Extract the content from the Word document
content = []
for paragraph in doc.paragraphs:
    content.append(paragraph.text)

# Remove empty rows from content
content = [row for row in content if row.strip()]

# Define column names
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','class']

# Create a dictionary of column names and content
data = {}
for i, col_name in enumerate(column_names):
    data[col_name] = [row.split(',')[i].strip() if len(row.split(',')) > i else '' for row in content]

# Create a DataFrame from the dictionary
test_df = pd.DataFrame(data)

# Define the output CSV file path
csv_file_path = 'test.csv'

# Write the DataFrame to a CSV file
test_df.to_csv(csv_file_path, index=False)

# Read the CSV file
test_df = pd.read_csv(csv_file_path)

print("CSV file created successfully!")
```

```python
from sklearn.preprocessing import LabelEncoder

class DecisionTree:
    def __init__(self):
        self.tree = None

    def get_rules(self, feature_names, label_mapping, tree=None, rule=None):
        if tree is None:
            tree = self.tree

        if rule is None:
            rule = ""

        if isinstance(tree, dict):
            if rule:
                rule += " AND "

            feature_index = tree['feature_index']
            threshold = tree['threshold']
            feature_name = feature_names[feature_index]
            rule += f"{feature_name} <= {threshold}"

            left_rule = self.get_rules(feature_names, label_mapping, tree['left_tree'], rule)
            right_rule = self.get_rules(feature_names, label_mapping, tree['right_tree'], rule)

            rules = []
            if isinstance(tree['left_tree'], dict):
                rules.extend(left_rule)
            else:
                class_label = label_mapping[tree['left_tree']]
                rules.append((left_rule, class_label))

            if isinstance(tree['right_tree'], dict):
                rules.extend(right_rule)
            else:
                class_label = label_mapping[tree['right_tree']]
                rules.append((right_rule, class_label))

            return rules
        else:
            class_label = label_mapping[tree]
            return [(rule, class_label)]


    def calculate_entropy(self, y):
        classes, counts = np.unique(y, return_counts=True)
        probabilities = counts / len(y)
        entropy = -np.sum(probabilities * np.log2(probabilities))
        return entropy

    def calculate_information_gain(self, X, y, feature_index, threshold):
        left_indices = np.where(X[:, feature_index] <= threshold)[0]
        right_indices = np.where(X[:, feature_index] > threshold)[0]

        left_y = y[left_indices]
        right_y = y[right_indices]

        parent_entropy = self.calculate_entropy(y)
        left_entropy = self.calculate_entropy(left_y)
        right_entropy = self.calculate_entropy(right_y)

        left_weight = len(left_y) / len(y)
        right_weight = len(right_y) / len(y)

        information_gain = parent_entropy - (left_weight * left_entropy) - (right_weight * right_entropy)
        return information_gain

    def choose_best_split(self, X, y):
        best_feature_index = None
        best_threshold = None
        best_information_gain = -np.inf

        for feature_index in range(X.shape[1]):
            unique_values = np.unique(X[:, feature_index])
            thresholds = (unique_values[:-1] + unique_values[1:]) / 2

            for threshold in thresholds:
                information_gain = self.calculate_information_gain(X, y, feature_index, threshold)

                if information_gain > best_information_gain:
                    best_information_gain = information_gain
                    best_feature_index = feature_index
                    best_threshold = threshold

        return best_feature_index, best_threshold

    def build_tree(self, X, y):
        if len(np.unique(y)) == 1:
            return np.unique(y)[0]

        if X.shape[0] == 0:
            return np.bincount(y).argmax()

        best_feature_index, best_threshold = self.choose_best_split(X, y)

        if best_feature_index is None or best_threshold is None:
            return np.bincount(y).argmax()

        left_indices = np.where(X[:, best_feature_index] <= best_threshold)[0]
        right_indices = np.where(X[:, best_feature_index] > best_threshold)[0]

        left_tree = self.build_tree(X[left_indices], y[left_indices])
        right_tree = self.build_tree(X[right_indices], y[right_indices])

        tree = {
            'feature_index': best_feature_index,
            'threshold': best_threshold,
            'left_tree': left_tree,
            'right_tree': right_tree
        }

        return tree

    def fit(self, X, y):
        self.tree = self.build_tree(X, y)

    def predict_sample(self, sample, tree):
        if isinstance(tree, dict):
            feature_index = tree['feature_index']
            threshold = tree['threshold']
            if sample[feature_index] <= threshold:
                if isinstance(tree['left_tree'], dict):
                    return self.predict_sample(sample, tree['left_tree'])
                else:
                    return tree['left_tree']
            else:
                if isinstance(tree['right_tree'], dict):
                    return self.predict_sample(sample, tree['right_tree'])
                else:
                    return tree['right_tree']
        else:
            return tree

#     def predict(self, X):
#         predictions = []
#         for sample in X:
#             prediction = self.predict_sample(sample, self.tree)
#             predictions.append(prediction)
#         return np.array(predictions)
    def predict(self, X, tree=None):
        if tree is None:
            tree = self.tree

        predictions = []
        for sample in X:
            prediction = self.predict_sample(sample, tree)
            predictions.append(prediction)
        return np.array(predictions)


    def get_leaf_nodes(self, tree):
        if isinstance(tree, dict):
            left_leaf_nodes = self.get_leaf_nodes(tree['left_tree'])
            right_leaf_nodes = self.get_leaf_nodes(tree['right_tree'])
            return left_leaf_nodes + right_leaf_nodes
        else:
            return [tree]

    def prune_tree(self, X_val, y_val):
        best_accuracy = self.calculate_accuracy(X_val, y_val)
        best_pruned_tree = self.tree

        leaf_nodes = self.get_leaf_nodes(self.tree)

        for leaf_node in leaf_nodes:
            pruned_tree = self.tree.copy()

            # Replace the leaf node with the majority class
            pruned_tree = self.replace_leaf_node(pruned_tree, leaf_node, np.bincount(y_val).argmax())

            accuracy = self.calculate_accuracy(X_val, y_val, pruned_tree)

            if accuracy >= best_accuracy:  # Use >= instead of >
                best_accuracy = accuracy
                best_pruned_tree = pruned_tree
            else:
                break  # Stop pruning if accuracy doesn't improve

        self.tree = best_pruned_tree


    def replace_leaf_node(self, tree, leaf_node, value):
        if isinstance(tree, dict):
            if tree['left_tree'] == leaf_node:
                tree['left_tree'] = value
            else:
                tree['left_tree'] = self.replace_leaf_node(tree['left_tree'], leaf_node, value)

            if tree['right_tree'] == leaf_node:
                tree['right_tree'] = value
            else:
                tree['right_tree'] = self.replace_leaf_node(tree['right_tree'], leaf_node, value)

        return tree

    def calculate_accuracy(self, X, y, tree=None):
        if tree is None:
            tree = self.tree

        y_pred = self.predict(X, tree)
        accuracy = (y == y_pred).mean()
        return accuracy
```
